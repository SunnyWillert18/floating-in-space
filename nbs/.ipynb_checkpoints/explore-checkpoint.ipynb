```python
import pandas as pd
import numpy as np
import scipy as sp
import plotly.express as px
```

```python
lost = pd.read_csv('../data/lost_things.csv')
tours = pd.read_csv('../data/tour_data.csv')
transport = pd.read_csv('../data/transport_data.csv')
activities = pd.read_csv('../data/activities_data.csv')
```

```python
# formatting columns to be nicer to work with
def format_names(df):
    for col in df.columns:
        df = df.rename(columns={col: col.replace(' ','_')})

    return df
```

```python
lost = format_names(lost)
tours = format_names(tours)
transport = format_names(transport)
activities = format_names(activities)
```

```python
# coverting to datetimes
lost.month = pd.to_datetime(lost.month).dt.strftime('%m/%Y')
tours.month = pd.to_datetime(tours.date)#.dt.strftime('%m/%d/%Y')
```

```python
south_america = ['Colombia','Ecuador','Peru','Bolivia','Brasil']
central_america = ['Guatemala','Honduras','Nicaragua']

hours_in_a_day = 24
transport.country.unique()
```

### How much stuff has been lost?

```python
print ('number of things lost: {}'.format(lost.shape[0]))
print ('I lost things for many reasons, including: {}'.format(lost.reason.unique()))
print ('I left a lot of things behind on purpose: {} items'.format(lost[lost.reason == 'left'].shape[0]))
ripped = lost[lost.reason == 'ripped']
print ('I ripped a pair of {} {} times over {} months'.format(ripped.item.unique()[0], ripped.shape[0], ripped.month.unique().shape[0]))

cases = ['broken','stolen','lost']
lost_clip = lost[lost.reason.isin(cases)]
print ('\nTaking out these cases:')
for x in ['type', 'month', 'country','reason']:
    
    print ('The most things lost by {} were: {}'.format(x, lost_clip.groupby(x).item.count().idxmax()))

print ('and I was most chaotic with one item: {}'.format(lost_clip.groupby('item').type.count().idxmax()))
```

### Time to look at transportation

```python
ttt = transport.approximate_time.sum()
print ('I spent {} hours in total in different types of transport - {} days'.format(ttt, np.round(ttt/hours_in_a_day, 1)))
```

```python
for x in transport.type.unique():
    ct = transport[transport.type == x].shape[0]
    ctt = transport[transport.type == x].approximate_time.sum()
    print ('{} times in a {} -- {} hours'.format(ct, x, ctt))
```

#### Night bus suffering!!

```python
night_bus = transport[(transport.type.isin(['bus','van'])) & (transport.time_of_day == 'night')]

print ('{} buses'.format(night_bus.shape[0]))
print ('{} hours (~{} days())'.format(night_bus.approximate_time.sum(), np.round((night_bus.approximate_time.sum()/hours_in_a_day),1)))

night_bus_sa = night_bus[night_bus.country.isin(south_america)]
night_bus_ca = night_bus[night_bus.country.isin(central_america)]
print ('with {} buses in South America for {} hours total'.format(night_bus_sa.shape[0], night_bus_sa.approximate_time.sum()))
print ('with {} buses in Central America for {} hours total'.format(night_bus_ca.shape[0], night_bus_ca.approximate_time.sum()))
```
