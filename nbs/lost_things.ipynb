```python
import pandas as pd
import numpy as np
import scipy as sp
import plotly.express as px
import plotly.graph_objects as go
```

```python
# formatting columns to be nicer to work with
def format_names(df):
    for col in df.columns:
        df = df.rename(columns={col: col.replace(' ','_')})

    return df
```

```python
lost = pd.read_csv('../data/lost_things.csv')
lost = format_names(lost)

lost.head()
```

### How much stuff has been lost?

```python
print ('number of things lost: {}'.format(lost.shape[0]))
print ('I lost things for many reasons, including: {}'.format(lost.reason.unique()))
print ('I left a lot of things behind on purpose: {} items'.format(lost[lost.reason == 'left'].shape[0]))
ripped = lost[lost.reason == 'ripped']
print ('I ripped a pair of {} {} times over {} months'.format(ripped.item.unique()[0], ripped.shape[0], ripped.month.unique().shape[0]))

cases = ['broken','stolen','lost']
lost_clip = lost[lost.reason.isin(cases)]
print ('\nTaking out these cases:')
for x in ['type', 'month', 'country','reason']:
    
    print ('The most things lost by {} were: {}'.format(x, lost_clip.groupby(x).item.count().idxmax()))

print ('and I was most chaotic with one item: {}'.format(lost_clip.groupby('item').type.count().idxmax()))
```

```python
print ('In total, it cost me approximately ${}'.format(lost.approx_cost.sum()))
for x in lost.reason.unique():
    clip = lost[lost.reason == x]
    print ('The approx cost of {} things was ${} (highest: {})'.format(x, clip.approx_cost.sum(), clip[clip.approx_cost == clip.approx_cost.max()].item.values[0]))
```
