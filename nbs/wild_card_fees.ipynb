### Wild Card Fees

```python
import pandas as pd
import numpy as np
import scipy as sp
import itertools
import datetime
import plotly.express as px
import plotly.graph_objects as go
```

```python
wild_card_price = 4160
rand_value = 13
```

```python
wild_card = pd.read_csv('../data/wild_card_fees.csv')
wild_card = wild_card.rename(columns={'time (days)':'time_in_days'})
wild_card
```

```python
print ('The cost of the wild card was {}R (~${}), and I would have spent {}R (~${}) without it'.format(
   wild_card_price, wild_card_price/rand_value, wild_card.total.sum(), np.round(wild_card.total.sum()/rand_value,2)
))
print ('Total days spent in parks: {}'.format(wild_card.time_in_days.sum()))
print ('Average price of park entrance per day: {}R (~${})'.format(wild_card.cost.mean(), np.round(wild_card.cost.mean()/rand_value,2)))
```
